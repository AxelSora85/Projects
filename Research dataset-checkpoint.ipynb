```python
#---------Preprocessing---------
import nltk
from nltk.tokenize import word_tokenize
word_tokenizer = nltk.tokenize.WhitespaceTokenizer()
```

```python
#lemmatizing the dataset for method 1
lemmatizer = nltk.stem.WordNetLemmatizer()
def lemmatize_text(text):
    return [lemmatizer.lemmatize(w) for w in word_tokenizer.tokenize(text)]

#X_train = X_train.str.replace('[^\w\s]','')
X_train = X_train.apply(lemmatize_text)
X_test = X_test.apply(lemmatize_text)

X_train=X_train.str.join(" ")
X_test=X_test.str.join(" ")
```

```python
#Remove stopwords with the help of the nltk library stopwords. Import words to compare against and remove.
#relates to method 4
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
X_train = X_train.apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
X_test= X_test.apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
```

```python
#Lowercase preprocessing method 2
def lowerCase(t):
    return t.lower()

X_train1 = []
X_test1 = []

#train and test
for t in X_train:
    text = lowerCase(t)
    X_train1.append(text)
    
for t in X_test:
    text = lowerCase(t)
    X_test1.append(text) 
```

```python
#Preprocessing method 3  Puncuation
X_train1 = X_train.map(lambda x: x.strip('’#$%@'))
X_test1 = X_test.map(lambda x: x.strip('’#$%@'))
```

```python
import string #use method string.punctuation from string class part of method 3
X_train1 = [''.join(char for char in strings if char not in string.punctuation) for strings in X_train1]
X_train1
```

```python
X_test1 = [''.join(char for char in strings if char not in string.punctuation) for strings in X_test1]
X_test1
```

```python
#stop word removal for train and test set.Method 4
from nltk.corpus import stopwords
stop = stopwords.words('english')
from nltk.tokenize import word_tokenize
X_test2 = []

for x in X_test1:
    textTokens = word_tokenize(x)#break apart string into tokens
    tokensWithoutSw = [word for word in textTokens if not word in stop]#remove words not in stop
    fin = ' '.join(tokensWithoutSw)#reconnect strings back togeather
    X_test2.append(fin)#append finished data to list
        

X_test2
```
